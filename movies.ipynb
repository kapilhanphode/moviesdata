```python
#imported pandas package
import pandas as pd
```

```python
#store all data to variable df 
#dataset is present in local system
df=pd.read_csv('netflix_titles.csv')
df.head()
```

```python
#from type column selected only Movies
movies=df['type']=='Movie'
movies
```

```python
#displayed how dataset will look like with type=Movies
df1=df[movies]
df1.head()
```

```python
#from all dataset selected the required fields i.e title,director,cast,release_year
df2=df1[['title','director','cast','release_year']]
df2=df2.reset_index(drop=True)
```

```python
df2
```

```python
#drop all row and columns which contain Nan values
df2=df2.dropna()
```

# Fetching all the data from the database (MySql)

```python
#import package mysql connector
import mysql.connector as connector 
```

```python
#making the connection with the database mysql
con=connector.connect(host='localhost',user='root',passwd='Admin')
con
```

```python
#fetching all the data from the database to our local system in the form of csv
query2='select * from movies.final_movie_dataset'
cur=con.cursor()
cur.execute(query2)
with open ('masterfile.csv','w',encoding='utf-8') as data:
    header='title,director,cast,release_year\n'
    data.write(header)
    for i in cur:
        print(i)
        data.write(str(i))
        
```
